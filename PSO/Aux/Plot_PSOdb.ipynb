```python
# ---------------------------Plot one db-------
# Notes, PPI is FI. The notation FI is used in the manuscript

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

def plot_best_ppi(db_path):
    # Connect to the database
    conn = sqlite3.connect(db_path)
    
    # Load data from the "particles" table
    particles_df = pd.read_sql("SELECT * FROM particles;", conn)
    
    # Close the database connection
    conn.close()
    
    # Filter out infinite PPI values
    filtered_best_ppi_per_iteration = particles_df[particles_df["ppi"] != float('inf')].groupby("iteration")["ppi"].max()
    
    # Create a DataFrame for best PPI per iteration (ignoring Inf values)
    filtered_best_ppi_df = filtered_best_ppi_per_iteration.reset_index()
    filtered_best_ppi_df.columns = ["Iteration", "Best PPI"]
 
    # Plot Best PPI per Iteration (Ignoring Inf values)
    plt.figure(figsize=(10, 5))
    plt.plot(filtered_best_ppi_df["Iteration"], filtered_best_ppi_df["Best PPI"], linestyle='-', color='blue', alpha=0.7)
    plt.xlabel("Iteration")
    plt.ylabel("Best FI")
    plt.title("Best FI over iterations")
    plt.ylim(0, 1)  # Clip y-axis to range 0 - 1
    plt.xlim(0,250)
    plt.grid(True)
    plt.savefig('PSO1_plot.png',dpi=300, bbox_inches='tight')
    
    # Show the plot
    plt.show()

# Load db file according to path
db_path = "/Users/......./pso_r_7.db"
plot_best_ppi(db_path)
```

```python
# ------------------------------Plot several dbs-----------------

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
import os

def plot_multiple_best_ppi(db_paths):
    plt.figure(figsize=(10, 5))
    colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown']  # Define colors for different DBs
    
    for i, db_path in enumerate(db_paths):
        # Connect to the database
        conn = sqlite3.connect(db_path)
        
        # Load data from the "particles" table
        particles_df = pd.read_sql("SELECT * FROM particles;", conn)
        
        # Close the database connection
        conn.close()
        
        # Filter out infinite PPI values
        filtered_best_ppi_per_iteration = particles_df[particles_df["ppi"] != float('inf')].groupby("iteration")["ppi"].max()
        
        # Create a DataFrame for best PPI per iteration (ignoring Inf values)
        filtered_best_ppi_df = filtered_best_ppi_per_iteration.reset_index()
        filtered_best_ppi_df.columns = ["Iteration", "Best PPI"]
        
        # Smooth the plot using spline interpolation
        x = filtered_best_ppi_df["Iteration"].values
        y = filtered_best_ppi_df["Best PPI"].values
        
        if len(x) > 3:  # Ensure enough points for smoothing
            x_smooth = np.linspace(x.min(), x.max(), 300)
            spline = make_interp_spline(x, y, k=3)
            y_smooth = spline(x_smooth)
        else:
            x_smooth, y_smooth = x, y  # No smoothing if not enough points
        
        # Plot each dataset with different colors
        # plt.plot(x_smooth, y_smooth, linestyle='-', color=colors[i % len(colors)], alpha=0.7, label=os.path.basename(db_path))
        plt.plot(x_smooth, y_smooth, linestyle='-', color=colors[i % len(colors)], alpha=0.7)
        
        #plt.scatter(x, y, color=colors[i % len(colors)], marker='o', alpha=0.5)
    
    # Labels and title
    plt.xlabel("Iteration")
    plt.ylabel("Best FI")
    plt.title("Best FI Over Iterations")
    plt.ylim(0, 1)  # Clip y-axis to range 0 - 1
    plt.xlim(0, 100)
    plt.grid(True)
    plt.legend()
    plt.savefig('PSO_plot.png',dpi=300, bbox_inches='tight')
    
    # Show the plot
    plt.show()

# Load db files (current directory):
db_paths = ["pso_r_7.db","pso_QLevyII_1.db","pso_Qgauss_2.db"] 
plot_multiple_best_ppi(db_paths)
```
